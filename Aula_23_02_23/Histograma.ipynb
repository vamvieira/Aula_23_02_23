```python
from matplotlib import pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("lenna.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Mostrando o histograma da imagem
plt.hist(img.ravel(),256,[0,256])
plt.show
```
